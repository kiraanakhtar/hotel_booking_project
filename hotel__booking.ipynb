```python
#Discovering the Heart of Food Choices: Insights from the 20th to 80th Percentile
```

```python


# Count the occurrences of each food type and sort them in descending order
# Calculate the 20th and 80th percentiles to set thresholds
# Filter the food counts to keep only values between the 20th and 80th percentiles
# Set figure size to be smaller
# Smaller figure size
# Define a custom palette with shades of blue
# Create a barplot with sorted values using the custom blue palette and add edgecolor for clarity
# Add a title and axis labels with a clearer font size
# Add value labels to each bar
# Remove spines (outline) for a cleaner look
# Optimize layout to prevent cutting off labels
# Show the plot
food_counts = df['food'].value_counts().sort_values(ascending=False)


lower_threshold = food_counts.quantile(0.20)
upper_threshold = food_counts.quantile(0.80)


filtered_food_counts = food_counts[(food_counts > lower_threshold) & (food_counts < upper_threshold)]


plt.figure(figsize=(8, 5))

custom_palette = ['darkblue', 'steelblue', 'lightblue', 'skyblue', 'deepskyblue']


sns.barplot(x=filtered_food_counts.values, y=filtered_food_counts.index, palette=custom_palette)

plt.title('Discovering the Heart of Food Choices: Insights from the 20th to 80th Percentile', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Food', fontsize=14)


for i, value in enumerate(filtered_food_counts.values):
    plt.text(value + 0.2, i, str(value), va='center', fontsize=12)


sns.despine()


plt.tight_layout()


plt.show()
```

```python
#Food Preferences by Count
```

```python
# Count the occurrences of each food type and sort them in descending order
# Set figure size to be smaller
# Smaller figure size for better visibility
# Define a custom palette with shades of blue
# Create a barplot with sorted values using the custom blue palette and add edgecolor for clarity
# Add a title and axis labels with a clearer font size
#Suggested title
# Add value labels to each bar
# Optimize layout to prevent cutting off labels
# Remove spines (outlines) for a cleaner look
# Show the plot


food_counts = df['food'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(8, 5))

custom_palette = ['darkblue', 'steelblue', 'lightblue', 'skyblue', 'deepskyblue']

sns.barplot(x=food_counts.values, y=food_counts.index, palette=custom_palette, edgecolor='black')

plt.title('Food Preferences by Count', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Food', fontsize=14)

for i, value in enumerate(food_counts.values):
    plt.text(value + 0.2, i, str(value), va='center', fontsize=12)

plt.tight_layout()

sns.despine()

plt.show()
```

```python
#Analyzing Hotel Cancellations: A Comparative Overview
```

```python

# Set a smaller figure size for better visibility
# Adjusted size
# Create countplot with specific colors (dark blue and light blue)
# Reduced bar width
# Add a title and axis labels with a clearer font size
 # Smaller font size
# Remove outlines (spines) for a cleaner look
# Add data labels
# Customize legend
# Optimize layout to prevent cutting off labels
# Adjusted font sizes
# Show the plot
plt.figure(figsize=(8, 4))

sns.countplot(data=df, x='hotel', hue='is_canceled', palette=['#1f77b4', '#aec7e8'], width=0.3)

plt.title('Number of Cancellations by Hotel', fontsize=14)
plt.xlabel('Hotel', fontsize=12)
plt.ylabel('Number of Reservations', fontsize=12)

sns.despine()

for p in plt.gca().patches:
    height = p.get_height()
    plt.gca().annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                       ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                       textcoords='offset points')

plt.legend(title='Is Canceled', labels=['No', 'Yes'], fontsize=10, title_fontsize=11)

plt.tight_layout()

plt.show()
```

```python

#Bivariate Analysis: ADR vs. Lead Time
```

```python

# Set the figure size to be smaller
# Define a custom palette with shades of dark blue and light blue
# Create the scatter plot with custom palette and without gridlines
# Add a title and axis labels
# Customize legend title
# Remove grid lines
# Remove spines (outlines) for a cleaner look
# Optimize layout to prevent clipping
# Show the plot
plt.figure(figsize=(6, 4))

custom_palette = ['darkblue', 'lightblue']

sns.scatterplot(data=df, x='lead_time', y='adr', hue='hotel', palette=custom_palette)

plt.title('Average Daily Rate (ADR) vs. Lead Time', fontsize=14)
plt.xlabel('Lead Time (Days)', fontsize=12)
plt.ylabel('Average Daily Rate (ADR)', fontsize=12)

plt.legend(title='Hotel')

plt.grid(False)

sns.despine()

plt.tight_layout()

plt.show()
```

```python
#Time Series Analysis: Reservations Over Time
```

```python
# Convert reservation status date to datetime
# Group by reservation date and count occurrences

# Set figure size to be smaller and more compact
# Adjusted size for better visibility
# Plot the reservations over time with shades of blue
# Light blue color
# Add a title and axis labels with smaller font sizes
# Smaller title font size
 # Smaller x-axis font size
# Smaller y-axis font size

# Rotate x-axis labels for better readability
# Remove outlines for a cleaner look
# Optimize layout to prevent clipping
# Show the plot
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])


reservations_per_day = df.groupby('reservation_status_date').size()

plt.figure(figsize=(10, 5))


plt.plot(reservations_per_day.index, reservations_per_day, color='skyblue')


plt.title('Reservations Over Time', fontsize=12)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Number of Reservations', fontsize=10)

plt.xticks(rotation=45)


sns.despine()


plt.tight_layout()


plt.show()
```
